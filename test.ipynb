```python
# ------------------------------
# EDA for Expense IIGF.csv
# ------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load Dataset
df = pd.read_csv("Expense IIGF.csv", encoding='latin1')   # Change encoding if needed

print("\n--- First 5 Rows ---")
print(df.head())

print("\n--- Data Shape (rows, columns) ---")
print(df.shape)

print("\n--- Columns ---")
print(df.columns)

print("\n--- Info ---")
print(df.info())

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Statistical Summary ---")
print(df.describe(include='all'))


# ------------------------------------
# CLEANING: Convert Date Column if Exists
# ------------------------------------
date_cols = ["Date", "date", "DATE"]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
        print(f"\nConverted {col} to datetime.")


# ------------------------------------
# VISUALIZATIONS
# ------------------------------------

# Bar chart – count of categories / types
cat_columns = ["Category", "Type", "Expense Type", "Item"]

for col in cat_columns:
    if col in df.columns:
        plt.figure(figsize=(6,5))
        df[col].value_counts().plot(kind='bar')
        plt.title(f"{col} Count")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.grid(False)
        plt.show()
        break


# Pie chart – expense distribution by category
for col in cat_columns:
    if col in df.columns:
        plt.figure(figsize=(6,6))
        df[col].value_counts().plot(kind='pie', autopct='%1.1f%%')
        plt.title(f"{col} Distribution")
        plt.ylabel("")
        plt.show()
        break


# Histogram – Amount column
amount_cols = ["Amount", "Cost", "Expense", "Value", "Price"]

for col in amount_cols:
    if col in df.columns:
        plt.figure(figsize=(6,5))
        plt.hist(df[col].dropna(), bins=15)
        plt.title(f"{col} Distribution")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.grid(False)
        plt.show()
        break


# Line chart – Expenses Over Time
for dcol in date_cols:
    if dcol in df.columns:
        if any(acol in df.columns for acol in amount_cols):

            amount_col = next(ac for ac in amount_cols if ac in df.columns)

            daily_expense = df.groupby(dcol)[amount_col].sum()

            plt.figure(figsize=(8,5))
            daily_expense.plot(kind='line')
            plt.title("Expenses Over Time")
            plt.xlabel("Date")
            plt.ylabel("Total Expense")
            plt.grid(True)
            plt.show()

        break


# Correlation Heatmap
numeric_df = df.select_dtypes(include=[np.number])

if not numeric_df.empty:
    plt.figure(figsize=(8,6))
    plt.imshow(numeric_df.corr(), cmap='viridis')
    plt.colorbar()
    plt.xticks(range(len(numeric_df.columns)), numeric_df.columns, rotation=90)
    plt.yticks(range(len(numeric_df.columns)), numeric_df.columns)
    plt.title("Correlation Heatmap")
    plt.show()

```
