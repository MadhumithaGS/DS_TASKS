```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = np.random.randint(1, 20, size=10)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y, marker='o', color='blue')
plt.title("Line Plot")

plt.subplot(2, 2, 2)
plt.bar(x, y, color='green')
plt.title("Bar Plot")

plt.subplot(2, 2, 3)
plt.scatter(x, y, color='red', s=50)
plt.title("Scatter Plot")

plt.subplot(2, 2, 4)
plt.barh(x, y, color='orange')
plt.title("Area Plot")

plt.tight_layout()
plt.show()
```

```python
data = np.random.randn(200)

plt.figure(figsize=(6, 5))
plt.hist(data, bins=20, edgecolor='black', color='orange')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 200 Random Normal Values")
plt.grid(False)
plt.show()
```
