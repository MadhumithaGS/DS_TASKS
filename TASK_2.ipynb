```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("Amazon Sale Report.csv", low_memory=False)

# Create a DataFrame for value counts
status_counts = df['Status'].value_counts().reset_index()

status_counts.columns = ['Status', 'Count']   # Rename columns

# Bar plot
status_counts.plot(kind='bar', x='Status', y='Count', figsize=(7,4))

plt.title("Order Status Count")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```
